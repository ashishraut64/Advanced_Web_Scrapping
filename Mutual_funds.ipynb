```python
import pandas as pd
import numpy as np
from numpy import nan
import re
```

```python
with open('Mutual_Funds1.html','r',encoding='utf-8') as f:
    html=f.read()
```

```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(html,'lxml')
```

```python
containers = soup.find_all('div',{'class':"caption schemebox-font"})
```

```python
a=[]
for i in range(0,len(containers)):
    if i%2:
        pass
    else:
        a.append(containers[i].find_all('p',class_='SNthumhead')[0].text)
```

```python
b=[]
for i in a:
    b.append(re.sub('\n','',i))
b
```

```python
#Extracting Fund Name
name=[]
for i in containers:
    try:
        name.append(i.find('p',{'class':'SNthumhead'}).find('a').text)
    except:
        name.append(np.nan)
```

```python
#Final Fund Name List
fund_name=[]
for i in name:
    if i in fund_name:
        pass
    else:
        fund_name.append(i)
        
```

```python
#Extracting NAV,category,typr,benchmark
category=[]
NAV=[]
type=[]
benchmark=[]
for i in range(0,len(containers)):
    if i%2:
        pass
    else:
        category.append(containers[i].find_all('th',class_='mfbox')[0].text)
        NAV.append(containers[i].find_all('th',class_='mfbox')[2].text)
        type.append(containers[i].find_all('th',class_='mfbox')[3].text)
        benchmark.append(containers[i].find_all('th',class_='mfbox')[4].text)
```

```python
containers[0].find('table',class_='table table-striped').find_all('td')[4].text
```

```python
#Extrcating Performance CAGR%
month_6=[]
year_1=[]
year_3=[]
for i in range(0,len(containers)):
    if i%2:
        pass
    else:
        month_6.append(containers[i].find('table',class_='table table-striped').find_all('td')[3].text)
        year_1.append(containers[i].find('table',class_='table table-striped').find_all('td')[4].text)
        year_3.append(containers[i].find('table',class_='table table-striped').find_all('td')[5].text)
        
```

```python
#Extracting Min_Invest,Exp_Ratio,SIP_Invest
min_invest=[]
Exp_Ratio=[]
SIP_invest=[]
for i in range(0,len(containers)):
    if i%2:
        min_invest.append(containers[i].find_all('th')[0].text.strip(':').strip())
        if "," in containers[i].find_all('th')[3].text.strip(':').strip():
            SIP_invest.append(containers[i].find_all('th')[3].text.strip(':').strip())
        else:
            SIP_invest.append('N/A')
        if '%' in containers[i].find_all('th')[4].text.strip(':').strip():
            Exp_Ratio.append(containers[i].find_all('th')[3].text.strip(':').strip())
        else:
            Exp_Ratio.append(containers[i].find_all('th')[4].text.strip(':').strip())
        
```

```python
#Extracting Ratings
x=[]
for i in range(0,len(containers)):
    if i%2:
        pass
    else:
        x.append(containers[i].find_all('th',class_="mfbox")[1].find('i'))
```

```python
x=str(x)
rating=re.findall('[0-9]{1}',x)
```

```python
#converting to float
NAV1=[]
for i in NAV:
    NAV1.append(re.sub(',{1}','',i))
NAV1 = [x.strip(' ') for x in NAV1]

min_invest1=[]
for i in min_invest:
    min_invest1.append(re.sub(',{1}','',i))
```

```python
df = pd.DataFrame({
    'Scheme': fund_name,
    'Category':category,
    'Type':type,
    'Benchmark':benchmark,
    'Rating':rating,
    'Net_Asset_Value(Rs.)':NAV1,
    'CAGR% 6 Months':month_6,
    'CAGR% 1 Year':year_1,
    'CAGR% 3 Year':year_3,
    'Min. Invest(Rs.)':min_invest1,
    'Exp. Ratio(%)':Exp_Ratio,
    'SIP Min. Inv.(Rs.)':SIP_invest,
    
})
```

```python
df['Net_Asset_Value(Rs.)']=df['Net_Asset_Value(Rs.)'].astype('float')
df['Min. Invest(Rs.)']=df['Min. Invest(Rs.)'].astype('float')
df['Rating']=df['Rating'].astype('int64')
```

```python
df.to_csv('Mutual_Funds.csv')
```
